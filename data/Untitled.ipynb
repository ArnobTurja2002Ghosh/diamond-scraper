```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns


import ipywidgets as widgets
from ipywidgets import interact, interact_manual

df2=pd.read_csv('2020-11-29 12-26 PM.csv')
```

```python
df2
```

```python
df1=df2[df2['clarity']!='FL'].reset_index()
```

```python
df1
```

```python
df1 = df1.iloc[: , 1:]
```

```python
df1
```

```python
@interact(Origin=['All','Natural', 'Lab'])
def f(Origin):
  if(Origin=="All"):
    df11= df1.groupby(['color', 'clarity'])['carat'].mean().reset_index()
  else:
    df11=df1[df1['type']==Origin.lower()].groupby(['color', 'clarity'])['carat'].mean().reset_index()
  df_hm = df11.pivot(index='color', columns='clarity', values='carat').fillna(0)
  df_hm1=df_hm[['SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']]
  sns.heatmap(data=df_hm1, cmap='mako')
```
